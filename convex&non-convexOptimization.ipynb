1) $$sin(x) - x^2$$
2) $$sin(x) + cos(x)$$
3) $$x^4 + x^3 + x + 2$$
4) $$x^2 - x^4 + 2$$

```python
# Check if functions above are convex or not by plotting them.
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy

# 1
x = np.arange(-10, 10, 0.1)
y = np.sin(x) - x**2
plt.title('sin(x) - x^2')
plt.plot(x,y)
plt.show()

# 2
x = np.arange(-10, 10, 0.1)
y = np.sin(x) + np.cos(x)
plt.title('sin(x) + cos(x)')
plt.plot(x,y)
plt.show()

# 3
x = np.arange(-5, 5, 0.1)
y = x**4 + x**3 + x + 2
plt.title('x^4 + x^3 + x + 2')
plt.plot(x,y)
plt.show()

# 4
x = np.arange(-1, 1, 0.01)
y = x**2 - x**4 + 2
plt.title('x^2 - x^4 + 2')
plt.plot(x,y)
plt.show()
```

```python
# Check if functions above are convex or not by using convex optimization.
from scipy.optimize import minimize 

def check_convexity_by_minimize(func, starting_point, run=2):
    for i in range(run):
        result = minimize(func, starting_point[i])
        solution = result['x']
        evaluation = func(solution)
        print(f'{i+1}st minimization: {func.__name__}({solution}) = {evaluation}')
    print()

def f1(x):
    return np.sin(x) - x**2

def f2(x):
    return np.sin(x) + np.cos(x)

def f3(x):
    return x**4 + x**3 + x + 2

def f4(x):
    return x**2 - x**4 + 2

check_convexity_by_minimize(f1, [-50, 50])
check_convexity_by_minimize(f2, [-50, 50])
check_convexity_by_minimize(f3, [-50, 50])
check_convexity_by_minimize(f4, [-50, 50])
```

    <h3> استفاده از روش‌های بهینه‌سازی برای توابع غیرمحدب
    </h3>
    در قسمت قبل تعدادی تابع به شما داده شد و با استفاده از دو روش محدب بودن یا نبودن آنها را بررسی کردید. در این قسمت باید با استفاده از روش <b>basin-hopping</b>
          <ul>
        <li>
  کمینه توابعی را که به عنوان توابع غیرمحدب دسته‌بندی کردید به دست آورید.
  </li>
  <li>
  از آنجایی که هیچ راهی وجود ندارد که basin-hopping متوجه پایان بهینه‌سازی شود، لازم است که تعداد مراحل اجرای الگوریتم را مشخص کنید. تاثیر این موضوع را در بهترشدن جواب بررسی کنید و بهترین تعداد مراحل اجرای الگوریتم را با استفاده از رسم نمودار پیدا کنید.
    </li>
  <li>
  مکان نقطه‌ی کمینه‌ی به دست آمده توسط این الگوریتم را روی نمودار تابع مشخص کنید. (به این ترتیب متوجه خواهید شد که الگوریتم شما چگونه عمل کرده است.)
    </li>
  </ul>
    </font>
</div>

```python
# Find the minimum of non-convex functions by the basin-hopping method. like before, you can only use scipy for this.
from scipy.optimize import basinhopping

def check_convexity_by_basinhopping(func, starting_point, no_itr=1000):
    result = basinhopping(func, starting_point, niter=no_itr)
    solution = result['x']
    evaluation = func(solution)
    print(f'{func.__name__}({solution}) = {evaluation}\n')
    x = np.arange(int(solution-5), int(solution+5), 0.1)
    y = [func(i) for i in x]
    plt.plot(x,y)
    plt.plot(solution, evaluation, marker='o')
    plt.show()

check_convexity_by_basinhopping(f1, -50)
check_convexity_by_basinhopping(f2, -50)
check_convexity_by_basinhopping(f4, -50)
```

<div align= "justify" dir="rtl">
<font face="XB Zar" size=3>
در تابع اول و سوم از توابع غیر محدب بالا چون مقدار مینیمم منفی بی نهایت است، پس این توابع با افزایش تعداد ایتریشن همچنان کم و کمتر میشوند و نقطه نشان داده شده در نمودار فقط مینیمم با تعداد ایتریشن 1000 را نشان می دهد.
تابع دوم نیز بسته به نقطه شروع ممکن است در هر کدام از مینیمم هایش گیر کند.
    </font>
</div>

```python
# Show the effect of the number of iterations on improving the answer.

def check_convexity_by_basinhopping(func, starting_point):
    x = []
    y = []
    iter = range(1000, 10001, 1000)
    for i in iter:
        print(f'Number of iteration: {i}')
        result = basinhopping(func, starting_point, niter=i)
        solution = result['x']
        evaluation = func(solution)
        print(f'{func.__name__}({solution}) = {evaluation}')
        x.append(solution)
        y.append(evaluation)
    plt.plot(iter, y)
    plt.show()
    y_min = min(y)
    y_min_index = y.index(y_min)
    print(f'The best minimum is {func.__name__}({x[y_min_index]}) = {y[y_min_index]} with {iter[y_min_index]} iteration.\n')

check_convexity_by_basinhopping(f1, -50)
check_convexity_by_basinhopping(f2, -50)
check_convexity_by_basinhopping(f4, -50)
```

 <b>پاسخ:</b> این متد ترکیبی از سه متد bisection، secant و inverse quadratic interpolation است و به این صورت عمل می کند که ابتدا دو مقدار ابتدا و انتهایی یک بازه را میگیرد و قضیه بولزانو را بررسی میکند که در آن بازه ریشه ای باشد، سپس با روش secant شروع می کند و و یک نقطه سوم بدست می آورد. پس از آن از روش inverse quadratic interpolation استفاده می کند تا با استفاده از این سه نقطه یک ریشه جدید بدست آورد، حال اگر این ریشه جدید بدست آمده خیلی منحرف شود، از روش bisection بجای interpolation استفاده می کند. با توجه به معیار هایی که دارد بعضی از ایتریشن ها را ریجکت می کند که در آن صورت در مرحله بعد از روش secant استفاده می کند.

<font face="XB Zar" size=3>
معادله زیر را در نظر بگیرید که جواب آن طبق شرط گذاشته شده، حتما باید مثبت و بین صفر و یک باشد.

یک بار آن را به کمک تابع 
root
 و یک بار دیگر هم به کمک همان متد 
 brent
  معادله را حل کنید و جواب‌های دو متد را با هم مقایسه کنید. 

  <b>پاسخ:</b> در استفاده از متد root در صورتی که نقطه شروع 0 باشد ریشه دیگر را که حدود 33- است پیدا می کند ولی با شروع از 0.99999 ریشه ای که پیدا می کند همان ریشه ای است که متد brentq پیدا می کند.
    </font>
</div>

```python
def obj_function(x, k=32.5):
    return x ** 2 / (1 - x) - k

x0 = 0.99999
x1 = 0
root_res = scipy.optimize.root(obj_function, x0)
print(f'Root method result with starting point {x0}: {root_res.x[0]}')
root_res = scipy.optimize.root(obj_function, x1)
print(f'Root method result with starting point {x1}: {root_res.x[0]}')

brentq_res = scipy.optimize.brentq(obj_function, 0, 0.99999)
print(f'Brent method result: {brentq_res}')
```

```python
# minimize the negative entropy to reach the best p
from scipy.special import comb
import numpy as np

def binomial_distribution(n, k, p):
    return comb(n, k) * (p**k) * ((1-p)**(n-k))

def negative_entropy(x):
    N = 5
    S = 0
    for i in range(N+1):
        p_i = binomial_distribution(N, i, x)
        S += p_i * np.log(p_i)
    return S

result = minimize(negative_entropy, 0.9, method='SLSQP')
solution = result['x']
evaluation = negative_entropy(solution)
print(f'Max entropy for p = {solution[0]} is {-evaluation[0]}')
```

```python
x = np.linspace(-2, 2, 100)

#make sure it's square
plt.figure(figsize=(6,6))

#have to plot the circle in two steps
plt.plot(x, np.sqrt(4 - x**2), color='blue')
plt.plot(x, -np.sqrt(4 - x**2), color='blue')

#now the point
plt.plot(5, 6, 'ro')

plt.xlim(-7, 7)
plt.ylim(-7, 7)

plt.show()
```

```python
# distance of two points
def dist(x1, x2):
    return np.sqrt((x1[0] - x2[0])**2 + (x1[1] - x2[1])**2)
```

```python
# our constraint function(x is a 2d tuple)
def circle(x):
    return x[0]** 2 + x[1]**2 - 4

# our objective function(x is a 2d tuple)
def objective(x):
    return dist((5, 6), x)

my_constraints = {'type': 'eq', 'fun': circle}
# use scipy.optimize.minimize to find the answer with given constraints and
# save answer in result variable
result = minimize(objective, (-2, 9), constraints=my_constraints)
```
